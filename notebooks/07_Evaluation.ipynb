## Retrieval Evaluation Metric For FinQA and ConvFinQA

### String Presence with Reference 
Calculates the per-gold context recall. Output shows what fraction of the gold/reference contexts that was matched by at least one of the retrieved results.

```python
from typing import List, Dict, Tuple, Iterable, Optional
import re
import numpy as np
import pandas as pd
from tqdm import tqdm
from math import log2
```

Helper functions: 

```python
def normalize_text(s: str) -> str:
    if not isinstance(s, str):
        return ""
    s = s.lower().strip()
    # collapse whitespace and common artifacts
    s = s.replace("\t", " ")
    s = s.replace("\n", " ")
    s = " ".join(s.split())
    return s

NUM_RE = re.compile(r"(?<![\w.])(?:\d{1,3}(?:,\d{3})*|\d+)(?:\.\d+)?%?")


def extract_numbers(text: str) -> list[str]:
    out = []
    for m in NUM_RE.findall(text or ""):
        out.append(m.replace(",", ""))  # keep % if present
    return out

def as_float_safe(s: str) -> Optional[float]:
    try:
        return float(s.rstrip("%")) if s.endswith("%") else float(s)
    except Exception:
        return None
    

def numbers_coverage(
    gold_nums: list[str],
    retr_nums: list[str],
    *,
    tol: float = 0.0,
    min_unique_hits: int = 4,
    min_coverage: float = 0.8,
) -> bool:
    """
    Returns True if the retrieved numerics cover the gold numerics well enough.
    - tol: absolute tolerance (e.g., 0.01)
    - min_unique_hits: at least this many distinct numeric values must match
    - min_coverage: matched / len(unique gold nums) threshold
    Matching is done by exact string OR numeric within tolerance.
    """
    if not gold_nums or not retr_nums:
        return False

    # dedupe to avoid inflated coverage from repeated values
    gold_unique = list(dict.fromkeys(gold_nums))
    retr_unique = list(dict.fromkeys(retr_nums))

    retr_set = set(retr_unique)  # fast string matches
    retr_floats = [as_float_safe(x) for x in retr_unique]
    retr_floats = [x for x in retr_floats if x is not None]

    hits = 0
    matched_values = set()

    for g in gold_unique:
        if g in retr_set:
            hits += 1
            matched_values.add(g)
            continue
        # try numeric tolerance
        gf = as_float_safe(g)
        if gf is None or tol <= 0 or not retr_floats:
            continue
        # any retrieved within tolerance counts
        for rf in retr_floats:
            if abs(gf - rf) <= tol:
                hits += 1
                matched_values.add(g)
                break

    coverage = hits / max(len(gold_unique), 1)
    return hits >= min_unique_hits and coverage >= min_coverage

def is_table_like(
    text: str,
    *,
    min_numbers: int = 8,
    min_num_density: float = 0.15,
) -> bool:
    """
    Heuristic: treat as table-like if it has lots of numbers and num-density is high.
    """
    if not text:
        return False
    nums = extract_numbers(text)
    if len(nums) < min_numbers:
        return False

    # crude density: numeric tokens / total tokens
    toks = re.findall(r"\S+", text)
    density = (len(nums) / max(len(toks), 1))
    return density >= min_num_density

def exact_or_table_numeric_match(
    gold: str,
    retrieved: str,
    *,
    use_table_numeric: bool,
    table_tol: float = 0.0,
    min_unique_hits: int = 4,
    min_coverage: float = 0.8,
) -> bool:
    """
    First try exact containment; if that fails and use_table_numeric=True,
    compare numeric coverage.
    """
    g = normalize_text(gold)
    r = normalize_text(retrieved)

    # exact containment (cheap & strict)
    if g and r and (g in r or r in g):
        return True

    if not use_table_numeric:
        return False

    g_nums = extract_numbers(gold)
    r_nums = extract_numbers(retrieved)
    return numbers_coverage(
        g_nums, r_nums,
        tol=table_tol,
        min_unique_hits=min_unique_hits,
        min_coverage=min_coverage,
    )
```

```python
def string_presence_with_reference(
    dataset: List[Dict],
    top_k_context_key: str = "retrieved_contexts",
    gold_context_key: str = "reference_contexts",
    k: int = 10,
    # table matching controls
    detect_tables: "auto|bool" = "auto",     # "auto" (heuristic), True (force), or False (disable)
    table_tol: float = 0.0,                  # numeric tolerance for tables (e.g., 0.01)
    min_unique_hits: int = 4,
    min_coverage: float = 0.8,
    # optional per-sample override key: sample["is_table_reference"] = True/False
    table_flag_key: str = "is_table_reference",
    verbose: bool = False,
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Exact-only evaluator with a numeric fallback for table-like references.

    - Per retrieved item, we mark relevant if:
        exact containment  OR  (reference is table-like AND numeric coverage passes)
    - Per gold recall uses the same logic.
    - You can force table behavior globally with detect_tables=True,
      disable with False, or auto-detect with "auto".
    - You can override per sample by setting sample[table_flag_key] = True/False.
    """
    recall_rows = []
    debug_rows = []

    for sample in tqdm(dataset):
        question = sample.get("user_input", "N/A")
        retrieved_raw = (sample.get(top_k_context_key, []) or [])[:k]
        gold_raw = sample.get(gold_context_key, []) or []

        retrieved_norm = [normalize_text(r) for r in retrieved_raw if isinstance(r, str)]
        gold_norm = [normalize_text(g) for g in gold_raw if isinstance(g, str)]

        # decide table-ness for this sample (per gold string)
        per_gold_is_table = []
        for g_raw in gold_raw:
            # per-sample override wins if present
            if table_flag_key in sample:
                is_tbl = bool(sample[table_flag_key])
            else:
                if detect_tables is True:
                    is_tbl = True
                elif detect_tables is False:
                    is_tbl = False
                else:  # "auto"
                    is_tbl = is_table_like(g_raw)
            per_gold_is_table.append(is_tbl)

        # per-retrieved flags (True if any gold deems it relevant)
        retrieved_hit_exact = []
        for r_raw in retrieved_raw:
            hit = False
            for g_raw, is_tbl in zip(gold_raw, per_gold_is_table):
                if exact_or_table_numeric_match(
                    g_raw, r_raw,
                    use_table_numeric=is_tbl,
                    table_tol=table_tol,
                    min_unique_hits=min_unique_hits,
                    min_coverage=min_coverage,
                ):
                    hit = True
                    break
            retrieved_hit_exact.append(hit)

        # per-gold recall + missing lists
        total = len(gold_norm)
        em_hits_gold = 0
        missing_em = []
        for g_raw, is_tbl in zip(gold_raw, per_gold_is_table):
            ok = any(
                exact_or_table_numeric_match(
                    g_raw, r_raw,
                    use_table_numeric=is_tbl,
                    table_tol=table_tol,
                    min_unique_hits=min_unique_hits,
                    min_coverage=min_coverage,
                )
                for r_raw in retrieved_raw
            )
            if ok:
                em_hits_gold += 1
            else:
                missing_em.append(normalize_text(g_raw))

        recall_em = em_hits_gold / total if total else 0.0

        recall_rows.append({
            "question_id": sample.get("question_id", "N/A"),
            "question": question,
            "reasoning_type": sample.get("reasoning_type", "N/A"),
            "total_gold": total,
            "retrieved_k": len(retrieved_norm),
            "recall_em": recall_em,
            "missing_em": missing_em,
            "source": sample.get("source", "N/A"),
        })

        debug_rows.append({
            "question_id": sample.get("question_id", "N/A"),
            "question": question,
            "reasoning_type": sample.get("reasoning_type", "N/A"),
            "retrieved_norm": retrieved_norm,
            "reference_norm": gold_norm,
            "retrieved_hit_exact": retrieved_hit_exact,  # used for nDCG
            "gold_table_like": per_gold_is_table,
            "source": sample.get("source", "N/A"),
        })

    recall_df = pd.DataFrame(recall_rows)
    debug_df = pd.DataFrame(debug_rows)
    return recall_df, debug_df
```

## Retrieval Evaluation Metrics For FinDER 

### Chunk Doc Substric Presence 

```python
from typing import List, Dict, Tuple
import re
import pandas as pd
from tqdm import tqdm
```

Helper Function: 

```python

_WS_RE = re.compile(r"\s+")
def normalize_text(s: str) -> str:
    if not isinstance(s, str):
        return ""
    # Lowercase, strip, collapse ALL whitespace to single spaces
    s = s.lower().strip()
    s = s.replace("\t", " ").replace("\n", " ")
    s = _WS_RE.sub(" ", s)
    return s
```

```python
def string_presence_doc_substring(
    dataset: List[Dict],
    top_k_context_key: str = "retrieved_contexts",   # list[str] (ranked chunks)
    gold_context_key: str = "reference_contexts",    # list[str] (full doc text or big blocks)
    k: int = 10,
    min_chunk_chars: int = 20,                       # ignore very short retrieved snippets
    verbose: bool = False,
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Exact-only evaluator (reversed):
      A retrieved chunk is relevant if its normalized text is a substring of ANY
      normalized reference context for that sample.

    Outputs:
      - recall_df: per-sample summary (hit@k, recall over gold docs as presence)
      - debug_df: per-rank boolean flags `retrieved_hit_exact` so you can reuse nDCG/MRR
    """
    recall_rows, debug_rows = [], []

    for sample in tqdm(dataset):
        question = sample.get("user_input", "N/A")
        retrieved_raw = (sample.get(top_k_context_key, []) or [])[:k]
        gold_raw = list(sample.get(gold_context_key, []) or [])

        # normalize once
        gold_norm = [normalize_text(g) for g in gold_raw]
        retrieved_norm = [normalize_text(r) for r in retrieved_raw]

        # per-retrieved: hit if retrieved_norm is a substring of any gold_norm
        retrieved_hit_exact = []
        match_into_doc_idx = []  # which reference index matched (for debugging)

        for r in retrieved_norm:
            if len(r) < min_chunk_chars:
                retrieved_hit_exact.append(False)
                match_into_doc_idx.append(None)
                continue

            hit = False
            matched_idx = None
            for gi, g in enumerate(gold_norm):
                if r and (r in g):
                    hit = True
                    matched_idx = gi
                    break
            retrieved_hit_exact.append(hit)
            match_into_doc_idx.append(matched_idx)

        # per-gold "recall": did any retrieved chunk appear inside that gold doc?
        total_gold = len(gold_norm)
        matched_doc_idxs = {idx for idx in match_into_doc_idx if idx is not None}
        recall_doc = (len(matched_doc_idxs) / total_gold) if total_gold else 0.0

        if verbose and not any(retrieved_hit_exact):
            print(f"[no-hit] {question[:80]}... (k={len(retrieved_norm)})")

        # --- summary row
        recall_rows.append({
            "question_id": sample.get("question_id", "N/A"),
            "question": question,
            "reasoning_type": sample.get("reasoning_type", "N/A"),
            "total_gold": total_gold,
            "retrieved_k": len(retrieved_norm),
            f"hit@{k}": 1 if any(retrieved_hit_exact) else 0,
            "recall_doc": recall_doc,
            "missing_em": [
                gold_norm[i] for i in range(total_gold) if i not in matched_doc_idxs
            ],
            "source": sample.get("source", "N/A"),
        })

        # --- debug row (what your nDCG/MRR use)
        debug_rows.append({
            "question_id": sample.get("question_id", "N/A"),
            "question": question,
            "reasoning_type": sample.get("reasoning_type", "N/A"),
            "retrieved_norm": retrieved_norm,
            "reference_norm": gold_norm,
            "retrieved_hit_exact": retrieved_hit_exact,   # <-- reuse for nDCG / MRR
            "matched_ref_index": match_into_doc_idx,      # which reference doc each rank hit
            "source": sample.get("source", "N/A"),
        })

    recall_df = pd.DataFrame(recall_rows)
    debug_df = pd.DataFrame(debug_rows)
    return recall_df, debug_df
```

## Retrieval Evaluation Metrics for All Datasets

### nDCG
Quality of the entire ranking, with more credit for relevant items near the top, but still rewards having multiple relevant results.

```python
def nDCG_with_reference(
    debug_df: pd.DataFrame,
    k: int = 10,
) -> pd.DataFrame:
    rows = []
    for _, row in debug_df.iterrows():
        question = row["question"]
        flags = list(row.get("retrieved_hit_exact", []) or [])[:k]

        if not flags:
            dcg = idcg = ndcg = 0.0
        else:
            dcg = sum((1 if f else 0) / log2(i + 2) for i, f in enumerate(flags))
            ideal = sorted((1 if f else 0 for f in flags), reverse=True)
            idcg = sum(rel / log2(i + 2) for i, rel in enumerate(ideal))
            ndcg = (dcg / idcg) if idcg > 0 else 0.0

        rows.append({
            "question_id": row["question_id"],
            "question": question,
            "reasoning_type": row.get("reasoning_type", "N/A"),
            f"dcg@{k}": dcg,
            f"idcg@{k}": idcg,
            f"ndcg@{k}": ndcg,
            "source": row.get("source", "N/A"),
        })

    ndcg_df = pd.DataFrame(rows)
    if not ndcg_df.empty:
        ndcg_df["mean_ndcg"] = ndcg_df[f"ndcg@{k}"].mean()
    return ndcg_df
```

### MRR
Similar as nDCG but only considers the first retrieved item. Great for examples where the task is to retrieve one relevant context. 

```python
def MRR_with_reference(
    debug_df: pd.DataFrame,
    k: int = 10,
) -> pd.DataFrame:
    rows = []
    for _, row in debug_df.iterrows():
        question = row["question"]
        flags = list(row.get("retrieved_hit_exact", []) or [])[:k]

        if not flags:
            rr = 0.0
        else:
            try:
                # Find first True and compute reciprocal rank
                first_hit_idx = flags.index(True)  # 0-based index
                rr = 1.0 / (first_hit_idx + 1)      # rank = index + 1
            except ValueError:
                rr = 0.0  # No True found

        rows.append({
            "question_id": row["question_id"],
            "question": question,
            "reasoning_type": row.get("reasoning_type", "N/A"),
            f"mrr@{k}": rr,
            "source": row.get("source", "N/A"),
        })

    mrr_df = pd.DataFrame(rows)
    if not mrr_df.empty:
        mrr_df["mean_mrr"] = mrr_df[f"mrr@{k}"].mean()
    return mrr_df
```

### Micro Precision, Micro Recall, Micro F1
These micro-averaged metrics evaluate retrieval performance by treating all retrieved items across all queries as one large set, rather than averaging per-query scores.<br>
**Micro Recall**<br>
Measures the fraction of all relevant items in the dataset that were successfully retrieved.<br>

Micro Recall = Total Relevant Retrieved/ Total Relevant Items <br>

​	
 
**Micro Precision**<br>
Measures the fraction of retrieved items that are actually relevant.<br>
Micro Precision = Total Relevant Retrieved/ Total Retrieved Items <br>

​	
 
**Micro F1**<br>
Harmonic mean of micro precision and micro recall, balancing both aspects.<br>
Micro F1 = (2⋅Precision ⋅ Recall)/(Precision + Recall) <br>

​	
 
**Why micro metrics?**
Unlike macro averaging (which computes metrics per query and then averages), micro metrics give a single overall score across the whole dataset.
They are especially useful when:<br>
- The number of relevant items per query is usually small, like 1.
- You want to capture the overall retrieval success rather than per-query balance.

```python
import pandas as pd

def _ensure_columns(
    recall_df: pd.DataFrame,
    debug_df: pd.DataFrame,
    dataset_name: str,
    k: int = 10,
) -> tuple[pd.DataFrame, pd.DataFrame]:
    r = recall_df.copy()
    d = debug_df.copy()

    # tag the source
    r["dataset"] = dataset_name
    d["dataset"] = dataset_name

    # unify recall column name
    if "recall_doc" not in r.columns and "recall_em" in r.columns:
        r["recall_doc"] = r["recall_em"]
    if "recall_em" not in r.columns and "recall_doc" in r.columns:
        r["recall_em"] = r["recall_doc"]

    # make sure totals exist
    if "total_gold" not in r.columns and "total_gold_docs" in r.columns:
        r["total_gold"] = r["total_gold_docs"]

    # recompute hit@k from debug flags if missing
    if f"hit@{k}" not in r.columns:
        # align by question and compute hit from debug flags
        hits_by_q = (
            d[["question", "retrieved_hit_exact"]]
            .assign(hit=lambda x: x["retrieved_hit_exact"].apply(lambda lst: int(bool(lst and any(lst)))))
            .set_index("question")["hit"]
        )
        r = r.set_index("question")
        r[f"hit@{k}"] = r.index.map(hits_by_q).fillna(0).astype(int)
        r = r.reset_index()

    # Ensure required columns exist even if empty
    for col in ["missing_em"]:
        if col not in r.columns:
            r[col] = [[] for _ in range(len(r))]

    return r, d


def combine_eval_frames(
    recall_df_others: pd.DataFrame, debug_df_others: pd.DataFrame,
    recall_df_find: pd.DataFrame,  debug_df_find: pd.DataFrame,
    k: int = 10,
):
    r1, d1 = _ensure_columns(recall_df_others, debug_df_others, "others", k=k)
    r2, d2 = _ensure_columns(recall_df_find,   debug_df_find,   "FinDER", k=k)

    recall_df_all = pd.concat([r1, r2], ignore_index=True)
    debug_df_all  = pd.concat([d1, d2], ignore_index=True)

    # (optional) keep only the unified columns you care about
    keep_cols_recall = ["question", "dataset", "total_gold", "retrieved_k",
                        f"hit@{k}", "recall_doc", "recall_em", "missing_em"]
    recall_df_all = recall_df_all[[c for c in keep_cols_recall if c in recall_df_all.columns]]

    return recall_df_all, debug_df_all
```

```python
def compute_micro_metrics(recall_df, recall_col="recall_doc"):
    """
    Compute micro precision, recall, and F1 from retrieval results.
    """
    if recall_df.empty:
        return {"precision_micro": 0.0, "recall_micro": 0.0, "f1_micro": 0.0}

    # Total relevant retrieved
    total_found = (recall_df[recall_col] * recall_df["total_gold"]).sum()
    
    # Total gold items
    total_gold = recall_df["total_gold"].sum()
    
    # Total retrieved
    total_retrieved = recall_df["retrieved_k"].sum()

    # Micro Recall
    recall_micro = total_found / total_gold if total_gold else 0.0

    # Micro Precision
    precision_micro = total_found / total_retrieved if total_retrieved else 0.0

    # Micro F1
    if precision_micro + recall_micro > 0:
        f1_micro = 2 * (precision_micro * recall_micro) / (precision_micro + recall_micro)
    else:
        f1_micro = 0.0

    return {
        "precision_micro": float(precision_micro),
        "recall_micro": float(recall_micro),
        "f1_micro": float(f1_micro)
    }
```

```python
import pandas as pd
from typing import Dict, List, Tuple, Callable, Optional

def evaluate_retrieval(
    df: pd.DataFrame,
    k: int = 10,
    *,
    presence_fn_map: Optional[Dict[str, Callable[[List[dict], int], Tuple[pd.DataFrame, pd.DataFrame]]]] = None,
    input_dataset_col: str = "source",
    merge_keys: Optional[List[str]] = None,
    recall_col_name: str = "recall_doc",
    # NEW: optional extra summary by reasoning type
    add_reasoning_breakdown: bool = True,          # NEW
    reasoning_col: str = "reasoning_type",         # NEW
) -> Dict[str, pd.DataFrame]:

    if input_dataset_col not in df.columns:
        raise ValueError(f"Expected '{input_dataset_col}' column in input df.")

    if presence_fn_map is None:
        presence_fn_map = {}

    datasets = df[input_dataset_col].dropna().astype(str).unique().tolist()

    all_recall: List[pd.DataFrame] = []
    all_debug:  List[pd.DataFrame] = []
    per_sample_chunks: List[pd.DataFrame] = []
    per_dataset_rows: List[pd.DataFrame] = []

    def _safe_mean(df_: pd.DataFrame, col: str) -> float:
        return float(df_[col].mean()) if (isinstance(df_, pd.DataFrame) and not df_.empty and col in df_.columns) else 0.0

    def _ensure_recall_schema(rec_df: pd.DataFrame, ds_name: str, k: int, recall_col_name: str) -> pd.DataFrame:
        df_ = rec_df.copy()

        # ensure dataset column
        if "dataset" not in df_.columns:
            df_["dataset"] = ds_name

        # ensure total_gold exists
        if "total_gold" not in df_.columns:
            if "reference_contexts" in df_.columns:
                df_["total_gold"] = df_["reference_contexts"].apply(
                    lambda x: len(x) if isinstance(x, (list, tuple)) else 1
                )
            else:
                df_["total_gold"] = 1

        # ensure recall indicator
        if recall_col_name not in df_.columns:
            alt = [f"hit@{k}", "hit@k", "hit", "found", "is_relevant", "recall"]
            found = next((c for c in alt if c in df_.columns), None)
            if found is not None:
                df_.rename(columns={found: recall_col_name}, inplace=True)
            elif "recall_em" in df_.columns:
                df_[recall_col_name] = df_["recall_em"]
            elif "missing_em" in df_.columns:
                def _derive_hit(row):
                    missing = row.get("missing_em", [])
                    try:
                        miss_n = len(missing)
                    except Exception:
                        miss_n = 0
                    total = row.get("total_gold", 1)
                    found_n = max(total - miss_n, 0)
                    return 1.0 if found_n > 0 else 0.0
                df_[recall_col_name] = df_.apply(_derive_hit, axis=1)
            else:
                raise KeyError(
                    f"Could not find recall indicator. Looked for '{recall_col_name}', "
                    f"{alt}, 'recall_em', or 'missing_em'. Columns present: {list(df_.columns)}"
                )

        # coerce to numeric 0/1
        df_[recall_col_name] = pd.to_numeric(df_[recall_col_name], errors="coerce").fillna(0).clip(0, 1)

        # convenience hit@k
        if f"hit@{k}" not in df_.columns:
            df_[f"hit@{k}"] = df_[recall_col_name]

        return df_

    def _infer_merge_keys(a: pd.DataFrame, b: pd.DataFrame) -> List[str]:
        """
        Choose merge keys that actually exist in *both* frames.
        Preference: explicit merge_keys -> id+dataset -> question+dataset -> dataset -> question -> any common.
        """
        # NEW: validate explicit merge_keys against both frames
        if merge_keys:
            cand = [k for k in merge_keys if k in a.columns and k in b.columns]
            if cand:
                return cand
        # try id fields
        for cand in ["question_id", "ID", "id"]:
            if cand in a.columns and cand in b.columns:
                keys = [cand]
                if "dataset" in a.columns and "dataset" in b.columns:
                    keys.append("dataset")
                return keys
        # question+dataset
        if "question" in a.columns and "question" in b.columns:
            if "dataset" in a.columns and "dataset" in b.columns:
                return ["question", "dataset"]
            return ["question"]
        # dataset only
        if "dataset" in a.columns and "dataset" in b.columns:
            return ["dataset"]
        # last resort
        common = [c for c in a.columns if c in b.columns]
        if common:
            return [common[0]]
        raise KeyError("No common merge keys between frames.")

    for ds in datasets:
        split = df[df[input_dataset_col] == ds]
        if split.empty:
            continue

        samples = split.to_dict(orient="records")
        fn = presence_fn_map.get(ds) or (
            string_presence_doc_substring if ds == "FinDER" else string_presence_with_reference
        )

        rec_df, dbg_df = fn(samples, k=k)

        # normalize recall schema and tag dataset
        rec_df = _ensure_recall_schema(rec_df, ds_name=ds, k=k, recall_col_name=recall_col_name)

        # make sure dbg_df also has a dataset column
        if "dataset" not in dbg_df.columns:
            dbg_df["dataset"] = ds

        # ranking metrics per dataset
        ndcg_df = nDCG_with_reference(dbg_df, k=k)   # expects column f"ndcg@{k}"
        mrr_df  = MRR_with_reference(dbg_df, k=k)    # expects column f"mrr@{k}"

        # micro metrics from recall_df
        micro = compute_micro_metrics(rec_df, recall_col=recall_col_name)

        # per-dataset summary row
        ds_summary = pd.DataFrame({
            "dataset": [ds],
            f"nDCG@{k}": [_safe_mean(ndcg_df, f"ndcg@{k}")],
            f"MRR@{k}":  [_safe_mean(mrr_df,  f"mrr@{k}")],
        })
        if isinstance(micro, pd.Series):
            for k_, v_ in micro.items():
                ds_summary[k_] = [v_]
        elif isinstance(micro, dict):
            for k_, v_ in micro.items():
                ds_summary[k_] = [v_]
        per_dataset_rows.append(ds_summary)

        # per-sample = debug + per-sample ndcg/mrr + recall cols
        per_sample = dbg_df.copy()

        key_ndcg = [c for c in [f"ndcg@{k}"] if isinstance(ndcg_df, pd.DataFrame) and c in ndcg_df.columns]
        key_mrr  = [c for c in [f"mrr@{k}"]  if isinstance(mrr_df,  pd.DataFrame) and c in mrr_df.columns]

        if key_ndcg:
            keys = _infer_merge_keys(per_sample, ndcg_df)
            per_sample = per_sample.merge(ndcg_df[keys + key_ndcg], on=keys, how="left")
        if key_mrr:
            keys = _infer_merge_keys(per_sample, mrr_df)
            per_sample = per_sample.merge(mrr_df[keys + key_mrr], on=keys, how="left")

        recall_cols = [c for c in rec_df.columns if c in {
            "dataset", recall_col_name, f"hit@{k}", "total_gold", "retrieved_k", "recall_em",
            reasoning_col, "question_id"   # NEW: keep helpful ids/labels if present
        }]
        keys = _infer_merge_keys(per_sample, rec_df)
        per_sample = per_sample.merge(rec_df[keys + [c for c in recall_cols if c not in keys]], on=keys, how="left")

        per_sample_chunks.append(per_sample)
        all_recall.append(rec_df)
        all_debug.append(dbg_df)

    # concat everything
    recall_all = pd.concat(all_recall, ignore_index=True) if all_recall else pd.DataFrame()
    debug_all  = pd.concat(all_debug,  ignore_index=True) if all_debug  else pd.DataFrame()
    per_sample_all = pd.concat(per_sample_chunks, ignore_index=True) if per_sample_chunks else pd.DataFrame()
    per_dataset_summary = pd.concat(per_dataset_rows, ignore_index=True) if per_dataset_rows else pd.DataFrame()

    # overall metrics
    overall_ndcg = nDCG_with_reference(debug_all, k=k) if not debug_all.empty else pd.DataFrame()
    overall_mrr  = MRR_with_reference(debug_all, k=k)  if not debug_all.empty else pd.DataFrame()
    overall_micro = compute_micro_metrics(recall_all, recall_col=recall_col_name) if not recall_all.empty else {}

    overall_summary = pd.DataFrame({
        "dataset": ["ALL"],
        f"nDCG@{k}": [_safe_mean(overall_ndcg, f"ndcg@{k}")],
        f"MRR@{k}":  [_safe_mean(overall_mrr,  f"mrr@{k}")],
    })
    if isinstance(overall_micro, pd.Series):
        for k_, v_ in overall_micro.items():
            overall_summary[k_] = [v_]
    elif isinstance(overall_micro, dict):
        for k_, v_ in overall_micro.items():
            overall_summary[k_] = [v_]

    # NEW: separate, additional summary by reasoning type (3 rows)
    per_reasoning_summary = pd.DataFrame()
    if add_reasoning_breakdown and not recall_all.empty and reasoning_col in recall_all.columns:
        rows = []
        # collect unique reasoning types present in the evaluated samples
        types = recall_all[reasoning_col].dropna().astype(str).unique().tolist()
        for rt in types:
            # slice per-type
            rec_sub = recall_all[recall_all[reasoning_col] == rt]
            dbg_sub = debug_all[debug_all[reasoning_col] == rt] if (not debug_all.empty and reasoning_col in debug_all.columns) else pd.DataFrame()

            # compute metrics for that slice
            ndcg_sub = nDCG_with_reference(dbg_sub, k=k) if not dbg_sub.empty else pd.DataFrame()
            mrr_sub  = MRR_with_reference(dbg_sub,  k=k) if not dbg_sub.empty else pd.DataFrame()
            micro_sub = compute_micro_metrics(rec_sub, recall_col=recall_col_name)

            row = {
                "reasoning_type": rt,
                f"nDCG@{k}": _safe_mean(ndcg_sub, f"ndcg@{k}"),
                f"MRR@{k}":  _safe_mean(mrr_sub,  f"mrr@{k}"),
            }
            if isinstance(micro_sub, pd.Series):
                row.update(micro_sub.to_dict())
            elif isinstance(micro_sub, dict):
                row.update(micro_sub)
            rows.append(row)

        per_reasoning_summary = pd.DataFrame(rows)

    # tidy orders
    def _order_dataset(df_):
        if df_.empty: return df_
        first = ["dataset"]
        ranking = [c for c in df_.columns if c.startswith("nDCG@") or c.startswith("MRR@")]
        others = [c for c in df_.columns if c not in first + ranking]
        return df_[first + ranking + others]

    def _order_reasoning(df_):
        if df_.empty: return df_
        first = ["reasoning_type"]
        ranking = [c for c in df_.columns if c.startswith("nDCG@") or c.startswith("MRR@")]
        others = [c for c in df_.columns if c not in first + ranking]
        return df_[first + ranking + others]

    # print summaries (datasets as before, then extra reasoning table)
    print("Per-dataset summaries:")
    print(_order_dataset(per_dataset_summary).to_string(index=False))
    if not per_reasoning_summary.empty:
        print("\nPer-reasoning_type summaries:")
        print(_order_reasoning(per_reasoning_summary).to_string(index=False))
    print("\nOverall summary:")
    print(_order_dataset(overall_summary).to_string(index=False))

    return {
        "per_sample": per_sample_all,
        "per_dataset_summary": _order_dataset(per_dataset_summary),
        "per_reasoning_summary": _order_reasoning(per_reasoning_summary),   # NEW
        "overall_summary": _order_dataset(overall_summary),
        "recall_all": recall_all,
        "debug_all": debug_all,
    }
```

```python
import time
import pandas as pd

def generate_answers_for_df(
    df: pd.DataFrame,
    gen,  # your ChatGPTGenerator instance
    *,
    question_col: str = "user_input",
    contexts_col: str = "retrieved_contexts",
    raw_col: str = "raw_model_output",
    max_context_chars: int = 12000,
    retries: int = 3,
    backoff: float = 2.0,
    show_progress: bool = True,
) -> pd.DataFrame:
    """Calls gen.generate(...) and stores the full raw text in `raw_col`."""
    def _as_list(x):
        if isinstance(x, list): return [str(t) for t in x]
        if isinstance(x, str):  return [x]
        return []

    it = df.iterrows()
    if show_progress:
        try:
            from tqdm import tqdm
            it = tqdm(it, total=len(df), desc="Generating (raw)")
        except Exception:
            pass

    out = df.copy()
    if raw_col not in out.columns:
        out[raw_col] = ""

    for idx, row in it:
        # skip if already populated
        if isinstance(out.at[idx, raw_col], str) and out.at[idx, raw_col].strip():
            continue

        q = str(row.get(question_col, "") or "")
        ctxs = _as_list(row.get(contexts_col, []))
        if ctxs:
            joined = "\n".join(ctxs)
            if len(joined) > max_context_chars:
                joined = joined[:max_context_chars]
            ctxs = [joined]

        last_err = None
        for attempt in range(1, retries + 1):
            try:
                raw = gen.generate(q, ctxs)        # <— your current generator; returns a string
                out.at[idx, raw_col] = raw.strip()
                last_err = None
                break
            except Exception as e:
                last_err = e
                time.sleep(backoff ** attempt)

        if last_err is not None:
            out.at[idx, raw_col] = ""
            print(f"[warn] row {idx}: generation failed after {retries} retries: {last_err}")

    return out
```

```python
import os
import torch
import pandas as pd
from typing import Iterable, List, Tuple
from transformers import AutoModelForSequenceClassification, AutoTokenizer

MODEL_NAME = "vectara/hallucination_evaluation_model"
BACKBONE_TOKENIZER = "microsoft/deberta-v3-base"

def _flatten_contexts(x) -> List[str]:
    out = []
    if x is None: return out
    if isinstance(x, str): return [x]
    if isinstance(x, dict):
        for v in x.values(): out.extend(_flatten_contexts(v))
        return out
    if isinstance(x, Iterable):
        for it in x: out.extend(_flatten_contexts(it))
        return out
    return [str(x)]

def _clean(s: str) -> str:
    return " ".join(str(s or "").split()).strip()

def _token_windows(text: str, tok, max_len: int, stride: int) -> List[str]:
    """Tokenize `text` and return overlapping windows decoded back to text."""
    ids = tok(text, add_special_tokens=False)["input_ids"]
    if not ids:
        return []
    if len(ids) <= max_len:
        return [text]
    step = max(1, max_len - stride)
    chunks = [ids[i:i+max_len] for i in range(0, len(ids), step)]
    return [tok.decode(c, skip_special_tokens=True) for c in chunks]

def build_hhem_model():
    # Use CPU by default to avoid MPS OOM. Switch to MPS/CUDA later if you want.
    os.environ.setdefault("PYTORCH_MPS_HIGH_WATERMARK_RATIO", "0.0")
    device = ("cpu")  # "mps" or "cuda" possible, but start with CPU (safer)
    model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_NAME, trust_remote_code=True, torch_dtype="auto"
    ).to(device).eval()
    return model

def add_faithfulness_hhem_predict_windowed(
    df: pd.DataFrame,
    *,
    question_col: str = "user_input",
    contexts_col: str = "retrieved_contexts",
    answer_col: str = "model_output",
    out_col: str = "faithfulness",
    include_question_in_premise: bool = False,
    # windowing (token counts; leave room if you later prepend question)
    ctx_tok_len: int = 350,
    ctx_stride: int = 150,
    # batching
    batch_size: int = 4,
) -> pd.DataFrame:
    """Scores faithfulness with HHEM via model.predict, using token windows and max aggregation."""
    tok = AutoTokenizer.from_pretrained(BACKBONE_TOKENIZER, use_fast=True)
    model = build_hhem_model()

    pairs: List[Tuple[str, str]] = []
    owners: List[int] = []  # row index per pair
    # Build pairs
    for idx, row in df.iterrows():
        ans = _clean(row.get(answer_col, ""))
        if not ans:  # if you want to score "I don't know", remove this guard
            continue

        ctxs = _flatten_contexts(row.get(contexts_col, []))
        ctxs = [c for c in ctxs if isinstance(c, str) and c.strip()]
        if not ctxs:
            continue

        ctx_joined = "\n".join(ctxs)
        # window the context by tokens
        windows = _token_windows(ctx_joined, tok, max_len=ctx_tok_len, stride=ctx_stride)
        if not windows:
            continue

        if include_question_in_premise:
            q = _clean(row.get(question_col, ""))
            prefix = f"QUESTION: {q}\nCONTEXT: "
        else:
            prefix = ""

        for w in windows:
            premise = _clean(prefix + w)
            if premise and ans:
                pairs.append((premise, ans))
                owners.append(idx)

    out = df.copy()
    out[out_col] = 0.0
    if not pairs:
        print("[HHEM] No valid pairs to score.")
        return out

    # predict in small batches; aggregate per row by max
    agg = {}
    pos = 0
    while pos < len(pairs):
        chunk = pairs[pos:pos+batch_size]
        with torch.inference_mode():
            try:
                scores = model.predict(chunk)
            except TypeError:
                scores = model.predict(chunk, batch_size=batch_size)
        scores = scores.tolist()
        for j, s in enumerate(scores):
            row_idx = owners[pos + j]
            agg[row_idx] = max(agg.get(row_idx, 0.0), float(s))
        pos += batch_size

    for i, val in agg.items():
        out.at[i, out_col] = val

    return out
```

```python
import re
import torch
import pandas as pd
from typing import Iterable, List, Tuple
from transformers import AutoModelForSequenceClassification, AutoTokenizer

MODEL_NAME = "vectara/hallucination_evaluation_model"
BACKBONE_TOKENIZER = "microsoft/deberta-v3-base"

def _flatten(x) -> List[str]:
    out=[]; 
    if x is None: return out
    if isinstance(x, str): return [x]
    if isinstance(x, dict):
        for v in x.values(): out += _flatten(v); return out
    try:
        for it in x: out += _flatten(it)
    except Exception:
        out=[str(x)]
    return out

def _clean(s: str) -> str:
    return " ".join(str(s or "").split()).strip()

_num = re.compile(r"^[+-]?\d+(?:\.\d+)?%?$")

def _make_claim(question: str, answer: str) -> str:
    """Turn (Q, A) into a natural-language claim for NLI."""
    q = _clean(question)
    a = _clean(answer)
    # If it's a bare number, add a % if the question implies percentage
    if _num.match(a) and "%" not in a and ("percent" in q.lower() or "%" in q):
        a = a + "%"
    # light rewrite: 'what was ... ?' -> '... was ...'
    claim_q = re.sub(r"\bwhat\s+was\b", "was", q, flags=re.I).rstrip("?")
    # generic claim
    return f"{claim_q} {a}" if claim_q else f"The answer is {a}."

def _token_windows(text: str, tok, max_len: int, stride: int) -> List[str]:
    ids = tok(text, add_special_tokens=False)["input_ids"]
    if not ids: return []
    if len(ids) <= max_len:
        return [text]
    step = max(1, max_len - stride)
    chunks = [ids[i:i+max_len] for i in range(0, len(ids), step)]
    return [tok.decode(c, skip_special_tokens=True) for c in chunks]

def add_faithfulness_hhem_predict_windowed(
    df: pd.DataFrame,
    *,
    question_col="user_input",
    contexts_col="retrieved_contexts",
    answer_col="model_output",
    out_col="faithfulness",
    evidence_col="faith_top_evidence",      # optional: see which window supported it
    ctx_tok_len=350, ctx_stride=150,
    batch_size=2, device="cpu",
):
    tok = AutoTokenizer.from_pretrained(BACKBONE_TOKENIZER, use_fast=True)
    model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_NAME, trust_remote_code=True, torch_dtype="auto"
    ).to(device).eval()

    out = df.copy()
    out[out_col] = 0.0
    if evidence_col: out[evidence_col] = ""

    for ridx, row in out.iterrows():
        ans = _clean(row.get(answer_col, ""))
        if not ans: 
            continue
        ctxs = [c for c in _flatten(row.get(contexts_col, [])) if isinstance(c, str) and c.strip()]
        if not ctxs:
            continue

        q = _clean(row.get(question_col, ""))
        claim = _make_claim(q, ans)
        ctx = "\n".join(ctxs)

        windows = _token_windows(ctx, tok, max_len=ctx_tok_len, stride=ctx_stride)
        if not windows: 
            continue

        # Build premise with question to enforce relevance
        premises = [f"QUESTION: {q}\nCONTEXT: {w}" for w in windows]

        best_s, best_w = 0.0, ""
        for i in range(0, len(premises), batch_size):
            chunk = [(prem, claim) for prem in premises[i:i+batch_size]]
            with torch.inference_mode():
                try:
                    s = model.predict(chunk)
                except TypeError:
                    s = model.predict(chunk, batch_size=batch_size)
            for prem, score in zip(premises[i:i+batch_size], s.tolist()):
                if float(score) > best_s:
                    best_s, best_w = float(score), prem

        out.at[ridx, out_col] = best_s
        if evidence_col: out.at[ridx, evidence_col] = best_w

    return out
```

### Faithfullness with LLM-As-A-Judge

```python
import json, re, time
from typing import List
import pandas as pd
from openai import OpenAI

# ---- small helpers ---------------------------------------------------------
def _flatten_ctx(x) -> List[str]:
    if x is None: return []
    if isinstance(x, str): return [x]
    if isinstance(x, dict):
        out=[]
        for v in x.values(): out += _flatten_ctx(v)
        return out
    try:
        out=[]
        for it in x: out += _flatten_ctx(it)
        return out
    except Exception:
        return [str(x)]

_STOP = set("""
a an the of in on for to and or if with from by at as is are was were be been being
what which who whose how many much percent percentage change value values outstanding
""".split())

def _terms(text: str) -> List[str]:
    toks = re.findall(r"[a-zA-Z][a-zA-Z\-]+", str(text).lower())
    return [t for t in toks if t not in _STOP and len(t) >= 3]

def _extract_json(s: str) -> dict:
    try:
        return json.loads(s)
    except Exception:
        m = re.search(r"\{.*\}", s, flags=re.S)
        if m:
            try:
                return json.loads(m.group(0))
            except Exception:
                return {}
        return {}

def add_faithfulness_llm_chunkwise(
    df: pd.DataFrame,
    *,
    client: OpenAI,
    model: str = "gpt-5-nano",
    question_col: str = "user_input",
    contexts_col: str = "retrieved_contexts",
    answer_col: str = "generated_answer",     # or "model_output"
    out_score_col: str = "faithfulness_llm",
    out_verdict_col: str = "faith_verdict",
    out_rationale_col: str = "faith_rationale",  # integer chunk index (1-based)
    out_error_col: str = "faith_error",          # full error text when things fail
    max_chunks_to_score: int = 6,
    require_question_terms: bool = True,
    per_chunk_char_limit: int = 1600,
    retries: int = 2, backoff: float = 1.4,
    one_based_index: bool = True,
) -> pd.DataFrame:
    """
    Nano-safe judge:
      • primary path: Responses API with `instructions` + `input`
      • fallback: Chat Completions with `developer` role
      • returns score∈[0,1], verdict, and best chunk index as rationale
      • captures the full error text in a separate column
    """

    INSTRUCTIONS = (
        "You are an evidence judge. Given a QUESTION, a single CONTEXT CHUNK, and an ANSWER, "
        "decide if this chunk alone is enough to support the answer (including any needed numbers). "
        'Return a JSON object: {"score": <0..1>, "verdict": "supported|partially_supported|unsupported"} . '
        "Return ONLY that JSON."
    )

    def _normalize_verdict(v: str) -> str:
        v = (v or "").strip().lower()
        if v in {"supported","unsupported","partially_supported"}:
            return v
        if "partial" in v: return "partially_supported"
        if "support" in v: return "supported"
        return "unsupported"

    def _ask_one(q: str, chunk: str, ans: str):
        user = (
            f"QUESTION:\n{q}\n\n"
            f"CONTEXT CHUNK:\n{chunk}\n\n"
            f"ANSWER:\n{ans}\n\n"
            "Reply with a single JSON object only."
        )
        last_err = None

        for a in range(retries):
            try:
                # --- 1) Preferred: Responses API (docs show this as primary) ---
                r = client.responses.create(
                    model=model,
                    instructions=INSTRUCTIONS,
                    input=user,
                )
                raw = getattr(r, "output_text", None) or str(r)
                obj = _extract_json(raw)
                s = max(0.0, min(1.0, float(obj.get("score", 0))))
                v = _normalize_verdict(obj.get("verdict", ""))
                return s, v, None
            except Exception as e1:
                last_err = e1
                # --- 2) Fallback: Chat Completions with 'developer' role ---
                try:
                    r2 = client.chat.completions.create(
                        model=model,
                        messages=[
                            {"role": "developer", "content": INSTRUCTIONS},
                            {"role": "user", "content": user},
                        ],
                    )
                    raw = r2.choices[0].message.content
                    obj = _extract_json(raw)
                    s = max(0.0, min(1.0, float(obj.get("score", 0))))
                    v = _normalize_verdict(obj.get("verdict", ""))
                    return s, v, None
                except Exception as e2:
                    last_err = e2
            time.sleep(backoff ** (a + 1))

        # both paths failed
        err_txt = f"{type(last_err).__name__}: {getattr(last_err, 'message', str(last_err))}"
        return 0.0, "unsupported", err_txt[:1200]

    out = df.copy()
    scores, verdicts, idxs, errs = [], [], [], []

    for _, row in out.iterrows():
        q = str(row.get(question_col, "") or "")
        ans = str(row.get(answer_col, "") or "").strip()
        chunks = [c for c in _flatten_ctx(row.get(contexts_col, [])) if isinstance(c, str) and c.strip()]

        if not chunks or not ans:
            scores.append(0.0); verdicts.append("unsupported")
            idxs.append(-1); errs.append("empty context or answer"); continue

        if require_question_terms:
            keys = _terms(q)
            if keys:
                kept = [c for c in chunks if any(k in c.lower() for k in keys)]
                if kept: chunks = kept

        chunks = chunks[:max_chunks_to_score]

        best_s, best_v, best_i, best_err = 0.0, "unsupported", -1, None
        for i, c in enumerate(chunks):
            s, v, err = _ask_one(q, c[:per_chunk_char_limit], ans)
            if err is not None and best_err is None:
                best_err = err  # keep first error for debugging
            if s > best_s:
                best_s, best_v, best_i = s, v, i

        scores.append(best_s)
        verdicts.append(best_v)
        idxs.append((best_i + 1) if one_based_index and best_i >= 0 else best_i)
        errs.append("" if best_i >= 0 else (best_err or ""))

    out[out_score_col] = scores
    out[out_verdict_col] = verdicts
    out[out_rationale_col] = idxs
    out[out_error_col] = errs
    return out
```

Faithfullness answers the following question: in what fraction of questions does my retriever fetch at least one chunk that supports the model’s answer?

Interpretation works as follows when combined with accuracy: <br>
High faithfulness & correct → retrieval brought the right evidence and model used it.<br>
High faithfulness & wrong → retrieval brought some supporting evidence, but it wasn’t actually the answer to the question (context mismatch / wrong doc section).<br>
Low faithfulness & wrong → retrieval didn’t surface support; model guessed.<br>
Low faithfulness & correct → model guessed the right number without evidence (risky).<br>

```python
def normalize_text(s: str) -> str:
    if s is None:
        return ""
    s = unicodedata.normalize("NFKC", s).lower()
    s = s.replace("–", "-").replace("—", "-")
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
from typing import List, Tuple, Dict, Any, Iterable
from collections import Counter, defaultdict
import math

DEFAULT_CONFIG = {
    "labels": {"correct": 0.8, "partial": 0.5},
    "arith": {"abs_tol": 1e-9, "rel_tol": 0.01, "max_rel": 1.0, "missing_penalty": 0.0},
    "non_num": {"stopwords": {
        "a","an","the","and","or","but","if","then","else","than","to","of","in","on","at","for","with","by",
        "is","are","was","were","be","been","being","as","that","this","these","those","it","its","from","into",
        "over","under","between","within","without","about","after","before","during","not","no","yes","do","does",
        "did","done","can","could","should","would","may","might","will","shall","i","you","he","she","we","they",
        "their","our","your","his","her","them","us"
    }},
    "comp": {"delims_regex": r"(?:,|;|\n| and | or |•|\u2022|\t)+", "min_item_len": 2}
}

def tokenize_content_words(s: str, stopwords: set) -> List[str]:
    s = normalize_text(s)
    tokens = re.findall(r"[a-z0-9][a-z0-9\-]*", s)
    norm = []
    for t in tokens:
        if t.endswith("s") and len(t) > 3:
            norm.append(t[:-1])
        else:
            norm.append(t)
    return [t for t in norm if t not in stopwords]

def lcs_len(a: List[str], b: List[str]) -> int:
    n, m = len(a), len(b)
    dp = [0]*(m+1)
    for i in range(1, n+1):
        prev = 0
        for j in range(1, m+1):
            tmp = dp[j]
            if a[i-1] == b[j-1]:
                dp[j] = prev + 1
            else:
                dp[j] = max(dp[j], dp[j-1])
            prev = tmp
    return dp[m]

def token_f1(ptoks: List[str], gtoks: List[str]) -> float:
    cp, cg = Counter(ptoks), Counter(gtoks)
    common = sum((cp & cg).values())
    prec = common / max(1, sum(cp.values()))
    rec  = common / max(1, sum(cg.values()))
    return 0.0 if prec + rec == 0 else 2*prec*rec/(prec+rec)

def extract_numbers(s: str) -> List[float]:
    if not s:
        return []
    s_norm = normalize_text(s)
    nums = []
    for m in re.finditer(r"(?P<sign>[-+])?\s*(?P<num>(?:\d+\.?\d*|\.\d+))(?:\s*(?P<pct>%))?", s_norm):
        val = float(m.group("num"))
        if m.group("sign") == "-":
            val = -val
        if m.group("pct"):  # treat 5% as 0.05
            val = val / 100.0
        nums.append(val)
    return nums

def align_numbers(pred_nums: List[float], gold_nums: List[float]):
    pairs, p, g = [], pred_nums[:], gold_nums[:]
    while p and g:
        best_i = best_j = None
        best_d = float("inf")
        for i, pv in enumerate(p):
            for j, gv in enumerate(g):
                d = abs(pv - gv)
                if d < best_d:
                    best_i, best_j, best_d = i, j, d
        pairs.append((p.pop(best_i), g.pop(best_j)))
    return pairs

def relative_error(pred: float, gold: float) -> float:
    return abs(pred-gold) if gold == 0.0 else abs(pred-gold) / max(1e-12, abs(gold))

def score_non_numerical(pred: str, gold: str, cfg: Dict[str, Any]) -> Tuple[float, Dict[str, Any]]:
    stop = cfg["stopwords"]
    ptoks = tokenize_content_words(pred, stop)
    gtoks = tokenize_content_words(gold, stop)
    f1 = token_f1(ptoks, gtoks)
    lcs = 0.0
    if ptoks and gtoks:
        lcs = lcs_len(ptoks, gtoks) / max(1, len(gtoks))
    return 0.5*f1 + 0.5*lcs, {"token_f1": f1, "lcs_recall": lcs}

def score_arithmetic(pred: str, gold: str, cfg: Dict[str, Any]) -> Tuple[float, Dict[str, Any]]:
    pnums, gnums = extract_numbers(pred), extract_numbers(gold)
    details = {"pred_nums": pnums, "gold_nums": gnums, "pair_scores": []}
    if not gnums:
        return score_non_numerical(pred, gold, DEFAULT_CONFIG["non_num"])
    if not pnums:
        return cfg["missing_penalty"], {**details, "reason": "no_numbers_in_pred"}
    pairs = align_numbers(pnums, gnums)
    scores = []
    for pv, gv in pairs:
        rerr = relative_error(pv, gv)
        if abs(pv-gv) <= cfg["abs_tol"] or rerr <= cfg["rel_tol"]:
            s = 1.0
        else:
            s = 1.0 / (1.0 + min(rerr, cfg["max_rel"]))  # 1 at 0 err, 0.5 at rel_err=1
        scores.append(s)
        details["pair_scores"].append({"pred": pv, "gold": gv, "rel_err": rerr, "score": s})
    unmatched = len(gnums) - len(pairs)
    if unmatched > 0:
        scores.extend([cfg["missing_penalty"]] * unmatched)
        details["unmatched_gold"] = unmatched
    return sum(scores)/len(scores), details

def split_items(s: str, delims_regex: str, min_item_len: int) -> List[str]:
    s_norm = normalize_text(s)
    parts = [p.strip() for p in re.split(delims_regex, s_norm) if p.strip()]
    return [p for p in parts if len(p) >= min_item_len]

def set_f1(pred_items: List[str], gold_items: List[str]) -> float:
    sp, sg = set(pred_items), set(gold_items)
    inter = len(sp & sg)
    prec = inter / max(1, len(sp))
    rec  = inter / max(1, len(sg))
    return 0.0 if prec + rec == 0 else 2*prec*rec/(prec+rec)

def score_compositional(pred: str, gold: str, cfg_comp: Dict[str, Any], cfg_non: Dict[str, Any]) -> Tuple[float, Dict[str, Any]]:
    pred_items = split_items(pred, cfg_comp["delims_regex"], cfg_comp["min_item_len"])
    gold_items = split_items(gold, cfg_comp["delims_regex"], cfg_comp["min_item_len"])
    phrase_f1 = set_f1(pred_items, gold_items) if gold_items else 0.0
    token_score, token_details = score_non_numerical(pred, gold, cfg_non)
    score = 0.6*phrase_f1 + 0.4*token_score
    return score, {"phrase_f1": phrase_f1, "token_component": token_details,
                   "pred_items": pred_items, "gold_items": gold_items}

def label_from_score(score: float, cfg: Dict[str, Any]) -> str:
    if score >= cfg["labels"]["correct"]:
        return "correct"
    if score >= cfg["labels"]["partial"]:
        return "partial"
    return "incorrect"

def score_answer(pred: str, gold: str, qtype: str, config: Dict[str, Any] = None) -> Dict[str, Any]:
    cfg = config or DEFAULT_CONFIG
    if qtype == "Arithmetic":
        score, details = score_arithmetic(pred, gold, cfg["arith"])
    elif qtype == "Non-numerical":
        score, details = score_non_numerical(pred, gold, cfg["non_num"])
    elif qtype == "Compositional":
        score, details = score_compositional(pred, gold, cfg["comp"], cfg["non_num"])
    else:
        score, details = score_non_numerical(pred, gold, cfg["non_num"])
    return {"score": max(0.0, min(1.0, score)),
            "label": label_from_score(score, cfg),
            "details": details}

def evaluate(dataset: Iterable[Dict[str, Any]], config: Dict[str, Any] = None) -> Dict[str, Any]:
    cfg = config or DEFAULT_CONFIG
    items = []
    sums, counts = 0.0, 0
    by_type = defaultdict(lambda: {"sum": 0.0, "n": 0})
    for ex in dataset:
        res = score_answer(ex["model_output"], ex["ground_truth_answer"], ex["reasoning_type"], cfg)
        items.append({**ex, **res})
        sums += res["score"]; counts += 1
        bt = by_type[ex["reasoning_type"]]; bt["sum"] += res["score"]; bt["n"] += 1
    mean = sums / counts if counts else 0.0
    by_type_mean = {k: (v["sum"]/v["n"] if v["n"] else 0.0) for k,v in by_type.items()}
    return {"mean": mean, "by_type": by_type_mean, "items": items}
```

```python
from typing import List, Tuple, Dict, Any, Iterable
from collections import Counter, defaultdict
import re, math

# ---- config ----
# ---- add to DEFAULT_CONFIG ----
DEFAULT_CONFIG = {
    "labels": {"correct": 0.8, "partial": 0.5},
    "arith": {"abs_tol": 1e-9, "rel_tol": 0.01, "max_rel": 1.0, "missing_penalty": 0.0},
    "non_num": {"stopwords": {
        "a","an","the","and","or","but","if","then","else","than","to","of","in","on","at","for","with","by",
        "is","are","was","were","be","been","being","as","that","this","these","those","it","its","from","into",
        "over","under","between","within","without","about","after","before","during","not","no","yes","do","does",
        "did","done","can","could","should","would","may","might","will","shall","i","you","he","she","we","they",
        "their","our","your","his","her","them","us"
    }},
    "routing": {
        "numeric_ratio_threshold": 0.40,  # >= 40% numeric -> treat as numerical
        "require_numeric_min": 1          # need at least 1 numeric token
    },
}


# ---- use your normalize_text (assumed defined elsewhere) ----
# def normalize_text(...): ...

# ---------- helpers ----------
def tokenize_content_words(s: str, stopwords: set) -> List[str]:
    s = normalize_text(s)
    tokens = re.findall(r"[a-z0-9][a-z0-9\-]*", s)
    norm = []
    for t in tokens:
        if t.endswith("s") and len(t) > 3:
            norm.append(t[:-1])
        else:
            norm.append(t)
    return [t for t in norm if t not in stopwords]

def lcs_len(a: List[str], b: List[str]) -> int:
    n, m = len(a), len(b)
    dp = [0]*(m+1)
    for i in range(1, n+1):
        prev = 0
        for j in range(1, m+1):
            tmp = dp[j]
            if a[i-1] == b[j-1]:
                dp[j] = prev + 1
            else:
                dp[j] = max(dp[j], dp[j-1])
            prev = tmp
    return dp[m]

def token_f1(ptoks: List[str], gtoks: List[str]) -> float:
    cp, cg = Counter(ptoks), Counter(gtoks)
    common = sum((cp & cg).values())
    prec = common / max(1, sum(cp.values()))
    rec  = common / max(1, sum(cg.values()))
    return 0.0 if prec + rec == 0 else 2*prec*rec/(prec+rec)

def extract_numbers(s: str) -> List[float]:
    if not s:
        return []
    s_norm = normalize_text(s)
    nums = []
    for m in re.finditer(r"(?P<sign>[-+])?\s*(?P<num>(?:\d+\.?\d*|\.\d+))(?:\s*(?P<pct>%))?", s_norm):
        val = float(m.group("num"))
        if m.group("sign") == "-":
            val = -val
        if m.group("pct"):  # 5% -> 0.05
            val = val / 100.0
        nums.append(val)
    return nums

def align_numbers(pred_nums: List[float], gold_nums: List[float]):
    pairs, p, g = [], pred_nums[:], gold_nums[:]
    while p and g:
        best_i = best_j = None
        best_d = float("inf")
        for i, pv in enumerate(p):
            for j, gv in enumerate(g):
                d = abs(pv - gv)
                if d < best_d:
                    best_i, best_j, best_d = i, j, d
        pairs.append((p.pop(best_i), g.pop(best_j)))
    return pairs

def relative_error(pred: float, gold: float) -> float:
    return abs(pred-gold) if gold == 0.0 else abs(pred-gold) / max(1e-12, abs(gold))

def _numeric_token_ratio(text: str, stopwords: set) -> float:
    """
    Ratio = numeric_tokens / (numeric_tokens + content_tokens)
    - numeric tokens: extracted by extract_numbers (handles %, signs, decimals)
    - content tokens: tokenize_content_words (stopwords removed)
    """
    if not text:
        return 0.0
    nums = extract_numbers(text)
    content = tokenize_content_words(text, stopwords)
    num_cnt = len(nums)
    den = num_cnt + len(content)
    return (num_cnt / den) if den > 0 else 0.0


# ---------- branch scorers ----------
def score_non_numerical(pred: str, gold: str, cfg: Dict[str, Any]) -> Tuple[float, Dict[str, Any]]:
    stop = cfg["stopwords"]
    ptoks = tokenize_content_words(pred, stop)
    gtoks = tokenize_content_words(gold, stop)
    f1 = token_f1(ptoks, gtoks)
    lcs = 0.0
    if ptoks and gtoks:
        lcs = lcs_len(ptoks, gtoks) / max(1, len(gtoks))
    return 0.5*f1 + 0.5*lcs, {"token_f1": f1, "lcs_recall": lcs}

def score_arithmetic(pred: str, gold: str, cfg: Dict[str, Any]) -> Tuple[float, Dict[str, Any]]:
    pnums, gnums = extract_numbers(pred), extract_numbers(gold)
    details = {"pred_nums": pnums, "gold_nums": gnums, "pair_scores": []}
    # If gold has no numbers, fall back to non-numerical (rare edge)
    if not gnums:
        return score_non_numerical(pred, gold, DEFAULT_CONFIG["non_num"])
    if not pnums:
        return cfg["missing_penalty"], {**details, "reason": "no_numbers_in_pred"}
    pairs = align_numbers(pnums, gnums)
    scores = []
    for pv, gv in pairs:
        rerr = relative_error(pv, gv)
        if abs(pv-gv) <= cfg["abs_tol"] or rerr <= cfg["rel_tol"]:
            s = 1.0
        else:
            s = 1.0 / (1.0 + min(rerr, cfg["max_rel"]))  # 1 at 0 err, 0.5 at rel_err=1
        scores.append(s)
        details["pair_scores"].append({"pred": pv, "gold": gv, "rel_err": rerr, "score": s})
    unmatched = len(gnums) - len(pairs)
    if unmatched > 0:
        scores.extend([cfg["missing_penalty"]] * unmatched)
        details["unmatched_gold"] = unmatched
    return sum(scores)/len(scores), details

# ---------- routing + evaluation ----------
def _route_branch(qtype: str, pred: str, gold: str, cfg: Dict[str, Any] = None) -> str:
    """
    Routing with numeric-token ratio:
    - Arithmetic/Numerical -> 'numerical'
    - Non-numerical        -> 'non_numerical'
    - Compositional/other  -> decide by ratio on gold (fallback to pred)
    """
    cfg = cfg or DEFAULT_CONFIG
    t = (qtype or "").strip().lower()

    # Explicit overrides
    if t in {"arithmetic", "numerical", "numeric", "number"}:
        return "numerical"
    if t in {"non-numerical", "non_numerical", "text", "verbal"}:
        return "non_numerical"

    # Heuristic for compositional / ambiguous
    basis = gold if (gold and gold.strip()) else pred
    ratio = _numeric_token_ratio(basis, DEFAULT_CONFIG["non_num"]["stopwords"])
    has_numeric = len(extract_numbers(basis)) >= DEFAULT_CONFIG["routing"]["require_numeric_min"]
    thr = DEFAULT_CONFIG["routing"]["numeric_ratio_threshold"]

    if has_numeric and ratio >= thr:
        return "numerical"
    return "non_numerical"


def label_from_score(score: float, cfg: Dict[str, Any]) -> str:
    if score >= cfg["labels"]["correct"]:
        return "correct"
    if score >= cfg["labels"]["partial"]:
        return "partial"
    return "incorrect"

def score_answer(pred: str, gold: str, qtype: str, config: Dict[str, Any] = None) -> Dict[str, Any]:
    cfg = config or DEFAULT_CONFIG
    branch = _route_branch(qtype, pred, gold, cfg)
    if branch == "numerical":
        score, details = score_arithmetic(pred, gold, cfg["arith"])
    else:
        score, details = score_non_numerical(pred, gold, cfg["non_num"])
    return {
        "score": max(0.0, min(1.0, score)),
        "label": label_from_score(score, cfg),
        "details": {**details, "branch": branch}
    }

def evaluate(dataset: Iterable[Dict[str, Any]], config: Dict[str, Any] = None) -> Dict[str, Any]:
    cfg = config or DEFAULT_CONFIG
    items = []
    sums, counts = 0.0, 0
    by_type = defaultdict(lambda: {"sum": 0.0, "n": 0})
    for ex in dataset:
        res = score_answer(ex["model_output"], ex["ground_truth_answer"], ex["reasoning_type"], cfg)
        items.append({**ex, **res})
        sums += res["score"]; counts += 1
        routed = _route_branch(ex["reasoning_type"], ex["model_output"], ex["ground_truth_answer"], cfg)
        by_type[routed]["sum"] += res["score"]; by_type[routed]["n"] += 1
    mean = sums / counts if counts else 0.0
    by_type_mean = {k: (v["sum"]/v["n"] if v["n"] else 0.0) for k,v in by_type.items()}
    return {"mean": mean, "by_type": by_type_mean, "items": items}
```

```python
# !pip install neo4j python-dotenv

import os
import re
import pandas as pd
from neo4j import GraphDatabase
from dotenv import load_dotenv

# 1) Load creds
load_dotenv()
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USERNAME = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
assert all([NEO4J_URI, NEO4J_USERNAME, NEO4J_PASSWORD]), "Missing Neo4j creds in .env"

# 2) Target in the graph
GRAPH_LABEL = "Document"
GRAPH_PROP  = "source_id"

def _safe(s: str) -> str:
    assert re.fullmatch(r"[A-Za-z_][A-Za-z0-9_]*", s), f"Unsafe label/prop: {s}"
    return s
GRAPH_LABEL = _safe(GRAPH_LABEL)
GRAPH_PROP  = _safe(GRAPH_PROP)

# 3) Pull keys from test_df
if "ID" not in test_df.columns:
    raise ValueError("test_df must contain a 'ID' column (e.g., 'EOG/2017/page_102.pdf').")

df = test_df.copy()
df["source_id_norm"] = df["ID"]

keys = df["source_id_norm"].tolist()

# 4) Check existence in Neo4j (chunked)
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))

def _check_batch(session, keys_batch):
    cypher = f"""
    UNWIND $keys AS k
    OPTIONAL MATCH (d:{GRAPH_LABEL} {{{GRAPH_PROP}: k}})
    RETURN k AS key, count(d) > 0 AS exists
    """
    res = session.run(cypher, keys=keys_batch)
    return {r["key"]: bool(r["exists"]) for r in res}

exists_map = {}
BATCH = 1000
with driver.session() as session:
    for i in range(0, len(keys), BATCH):
        batch = keys[i:i+BATCH]
        exists_map.update(_check_batch(session, batch))
driver.close()

# 5) Attach results + summary
df["in_kg"] = df["source_id_norm"].map(exists_map).fillna(False)

total = len(df)
present = int(df["in_kg"].sum())
missing = total - present
print(f"Checked (:Document {{{GRAPH_PROP}}}) for {total} rows.")
print(f"Found in KG: {present}  |  Missing: {missing}")

# Show a few missing for debugging
df.loc[~df["in_kg"], ["ID"]].head(10)
```

```python
def fill_reference_with_gold(df: pd.DataFrame) -> pd.DataFrame:
    """
    If reference_contexts is empty ([], None, or NaN) 
    and gold_context is not empty, copy gold_context into reference_contexts.
    """
    def _fix_row(row):
        ref = row.get("reference_contexts", None)
        gold = row.get("gold_context", None)

        # normalize emptiness
        is_ref_empty = (
            ref is None 
            or (isinstance(ref, float) and pd.isna(ref))
            or (isinstance(ref, list) and len(ref) == 0)
        )

        if is_ref_empty and gold not in (None, "", []):
            return gold
        return ref

    df = df.copy()
    df["reference_contexts"] = df.apply(_fix_row, axis=1)
    return df
```

```python
def normalize_gold_context(x):
    """
    Ensure gold_context is always a list of strings.
    """
    if isinstance(x, dict):
        # take just the values (strings)
        return list(x.values())
    elif isinstance(x, str):
        # wrap single string into a list
        return [x]
    elif isinstance(x, (list, tuple)):
        # already a list -> make sure all elements are strings
        return [str(el) for el in x]
    else:
        # unexpected type -> return empty list
        return []

# apply to your df
df_vector_formatted["reference_contexts"] = df_vector_formatted["reference_contexts"].apply(normalize_gold_context)
```

**Dataset-level differences**<br>
- For VectorRAG, latency is relatively consistent across datasets, clustering around ~1s with small variance.
- For HybridRAG, ConvFinQA queries are slightly faster than FinDER and FinQA, suggesting that dataset/query type complexity interacts with the boosting heuristics.

**Latency vs. Effectiveness**<br>
Correlations (Pearson & Spearman) are close to zero:<br>
→ Neither VectorRAG nor HybridRAG show a strong relationship between latency and effectiveness (nDCG/MRR).<br>
→ In other words, waiting longer doesn’t necessarily yield better results—retrieval quality is not latency-driven.<br>
This also implies HybridRAG’s higher latency is due to system overhead (reranking, heuristics, index lookups) rather than “deeper searching.”<br>

**Comparative insight**<br>
VectorRAG = efficiency-focused: consistent latency, reasonable performance, lower cost.<br>
HybridRAG = effectiveness-focused: higher retrieval accuracy at the cost of latency and stability.<br>
For interactive/real-time applications, VectorRAG might be preferable. For high-stakes accuracy tasks (e.g., finance QA), HybridRAG justifies the latency overhead.<br>
